# 6-level schema analysis (Cohen's Kappa (κ))

```python
from sklearn.metrics import cohen_kappa_score
annotator1 = df['tamara_tag']
annotator2 = df['katja_tag']
cohen_kappa_score(annotator1, annotator2)
```

# 3-level analysis (Cohen's Kappa (κ))

```python
df.loc[(df['tamara_tag'] == 'Negative') | (df['tamara_tag'] == 'M_Negative'), 'sent_tamara'] = 'Negative'
df.loc[(df['tamara_tag'] == 'Positive') | (df['tamara_tag'] == 'M_Positive'), 'sent_tamara'] = 'Positive'
df.loc[(df['tamara_tag'] == 'P_Neutral') | (df['tamara_tag'] == 'N_Neutral'), 'sent_tamara'] = 'Neutral'
```

```python
df.loc[(df['katja_tag'] == 'Negative') | (df['katja_tag'] == 'M_Negative'), 'sent_katja'] = 'Negative'
df.loc[(df['katja_tag'] == 'Positive') | (df['katja_tag'] == 'M_Positive'), 'sent_katja'] = 'Positive'
df.loc[(df['katja_tag'] == 'P_Neutral') | (df['katja_tag'] == 'N_Neutral'), 'sent_katja'] = 'Neutral'
```

```python
annotator1 = df['sent_tamara']
annotator2 = df['sent_katja']
cohen_kappa_score(annotator1, annotator2)
```
