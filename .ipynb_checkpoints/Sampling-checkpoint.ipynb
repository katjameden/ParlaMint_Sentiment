```python
import csv
import random
import pandas as pd
```

```python
meta = pd.read_csv("ParlaMint-SI-meta.csv", encoding='utf-8', delimiter=',')
text = pd.read_csv("ParlaMint-SI-text.csv", encoding='utf-8', delimiter='\t')
```

```python
df = pd.merge(meta, text, on='ID', how='left')
```

```python
df
df.to_csv("ParlaMint-SI-merged.csv", index=False)
```

# Sampling with corpus distribution

Sample each 300th utterance in the DataFrame, to land with 1038 samples. Fruther divide into two datasets: annotation sample (1000 samples) and the dry-run sample (38 samples).


```python
df_sample = df.iloc[::300, :]
df_sample
```

```python
df_sample_1000 = df_sample.iloc[:1000, :]
df_sample_38 = df_sample.iloc[1000:, :]
```

```python
df_1000_filtered = df_sample_1000[['ID', 'Text']]
df_test_filtered = df_sample_38[['ID', 'Text']]
```

```python
header_list= ['ID','Text','Annotation', 'Comments', 'Flag']

df_1000_filtered = df_1000_filtered.reindex(columns = header_list)
df_test_filtered = df_test_filtered.reindex(columns = header_list)
```

```python
df_1000_filtered
```

```python
df_test_filtered
```

```python
df_1000_filtered.to_csv("Datasets/ParlaMint_sample.csv", index = False)
df_test_filtered.to_csv("Datasets/ParlaMint_Tests_sample.csv", index = False)
```

# Development and test splits 

Splitting the full sample (ParlaMint_sample.csv, 1000 utterances) to the dev dataset (80%, 800 utterances) and test dataset (20%, 200 utterances), randomized.

```python
from sklearn.model_selection import train_test_split
test_size = 0.2  # 20% for test
dev_size = 0.8  # 80% for dev (implicit)

dev_data, test_data = train_test_split(df_1000_filtered, test_size=test_size, random_state=42)

dev_data
```

```python
test_data
```

```python
dev_data.to_csv("Datasets/ParlaMint_split_dev.csv", index=False)
```

```python
test_data.to_csv("Datasets/ParlaMint_split_test.csv", index=False)
```
