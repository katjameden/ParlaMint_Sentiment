```python
import pandas as pd
```

```python
df = pd.read_csv('../Datasets/Development_data.csv', encoding='utf-8')
df.head()
```

# Sentence sentiment score aggregation

Testing different aggregation methods (average, mean, etc), as well as investigating some key characteristics of the dataset (e.g., dispersion of the sentence sentiment scores within the utterance). 

## Mean and median

```python
aggregation = df.groupby('ID')['annotation_sent'].agg(
    count='count',
    mean ='mean', 
    median='median',
    min='min', 
    max='max', 
    std='std',
).reset_index()
```

```python
df = pd.merge(df, aggregation, on='ID', how='left')
df.head()
```

```python
#Checking the highest data dispersion by sorting the values by standard deviation score 
# count = df.sort_values(by = 'std', ascending=False)
#count.head()
```

```python
df['diff_mean'] = abs(df['annotation_utt_score'] - df['mean'])
df['diff_median'] = abs(df['annotation_utt_score'] - df['median'])

df.head()
```

```python
average_max = df.sort_values(by = 'diff_mean', ascending=False)
average_max.head(5)
```

```python
median_max = df.sort_values(by ='diff_median', ascending=False)
median_max.head(5)
```

```python
mean_diff = df['diff_mean'].mean()
median_diff = df['diff_median'].mean()

print(f"Average difference with mean: {mean_diff}")
print(f"Average difference with median: {median_diff}")
```

# Calculating differents weights and weighted average

The weights assigned determined by the following indicators:
<!-- This is currently just a test with contexts that could be important when considering sentiment in whole utterances -->
- the *position* of the sentence (e.g., if the sentence appears at the end of the utterance)
- the *length* of the sentences (the longer the sentence, the higher the weight?)
- the *intensity of the sentiment* - utilising the sentence sentiment scores to produce weighted sentiment score (using absolute values), to boost the extremities in the speeches. 

## Position

```python
df['position'] = df.groupby('ID').cumcount() + 1
df['position_max'] = df.groupby('ID')['position'].transform('max')
df.head()
```

```python
df['position_score'] = df['annotation_sent'] * df['position']
position_avg = df.groupby('ID').apply(lambda x: x['position_score'].sum() / x['position'].sum()).reset_index(name='position_avg')
position_avg
```

```python
df = df.merge(position_avg, on='ID')
df.head()
```

```python
df['diff_position'] = abs(df['annotation_utt_score'] - df['position_avg'])
position_avg = df['diff_position'].mean()
print("Average difference with intensity weights:", position_avg)
```

## Sentence length

```python
df['length'] = df['text_sent'].apply(lambda x: len(x.split()))
```

```python
aggregated_scores = (
    df.groupby("ID")
    .apply(lambda x: (x["annotation_sent"] * x["length"]).sum() / x["length"].sum())
    .reset_index(name="length_avg")
)
```

```python
df = df.merge(
    aggregated_scores,
    on="ID"
)

df.head()
```

```python
df['diff_length'] = abs(df['annotation_utt_score'] - df['length_avg'])
length_avg = df['diff_length'].mean()
print("Average difference with length weights:", length_avg)
```

## Sentiment intensity weighted average

```python
count_df = df.sort_values(by = 'annotation_sent', ascending=True)
count_df.head()
```

```python
df['intensity'] = df['annotation_sent'].abs()
df['intensity_score'] = df['annotation_sent'] * df['intensity']
df.head()
```

```python
intensity_avg = df.groupby('ID').apply(lambda x: x['intensity_score'].sum() / x['intensity'].sum()).reset_index(name='intensity_avg')
intensity_avg
```

```python
df = df.merge(intensity_avg, on='ID')
df.head()
```

```python
df['diff_intensity'] = abs(df['annotation_utt_score'] - df['intensity_avg'])
intensity_avg = df['diff_intensity'].mean()
print("Average difference with intensity weights:", intensity_avg)
```
