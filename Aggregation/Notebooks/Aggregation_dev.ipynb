```python
import pandas as pd
```

```python
df = pd.read_csv('../Datasets/Development_data.csv', encoding='utf-8')
df
```

```python
sent_mean = df.groupby('ID')['annotation_sent'].mean().reset_index()
```

```python
sent_avg = sent_mean.rename(columns={'annotation_sent':'avg_annotation'})
```

```python
sent_avg.reset_index(drop=True, inplace=True)
```

```python
df = pd.merge(df, sent_avg, on='ID', how='left')
df
```

```python
df.to_csv('../Datasets/Aggregation_testing.csv', encoding='utf-8', index=False)
```
