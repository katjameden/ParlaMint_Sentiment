# Sentence sentiment score aggregation

Testing different aggregation methods (average, mean, etc), as well as investigating some key characteristics of the dataset (e.g., dispersion of the sentence sentiment scores within the utterance). 

```python
aggregation = df.groupby('ID')['annotation_sent'].agg(
    count='count',
    mean ='mean', 
    median='median',
    min='min', 
    max='max', 
    std='std',
).reset_index()
```

## Calculating weights and weighted average

The weights assigned determined by the following indicators:
<!-- This is currently just a test with contexts that could be important when considering sentiment in whole utterances -->
- the *position* of the sentence (e.g., if the sentence appears at the end of the utterance)
- the *length* of the sentences (the longer the sentence, the higher the weight?)
- lower weights for the *procedural* sentences (maybe based on the “Chair” value for the utterance to signify high possibility of procedural language in sentences).
    - e.g. *if chair = True & (value ≠ Neutral_P |  value ≠ Neutral_N ) → lower weights set*

```python
df.head()
```

```python
df['position'] = df.groupby('ID').cumcount() + 1
#df['position_weight'] = df.groupby('ID')['position'].transform(lambda x: x.apply(lambda y: 1.5 if y == x.max() else 1 ))
df.head(3)
```

```python
df['position_max'] = df.groupby('ID')['position'].transform('max')
df
```

```python
def position_chair(row):
    chair = row['chair']
    position = row['position']
    max_position = row['position_max']
    if chair != True and position == max_position:
        return 1.5
    return 1

df['weight'] = df.apply(position_chair, axis=1)
df.drop(columns=['position_max'], inplace=True)
df.head(10)
```

```python
#check = df[df['chair'] == False]
#check
```

```python
df['weighted_score'] = df['annotation_sent'] * df['weight']
df['weighted_average'] = df.groupby('ID')['weighted_score'].transform('sum') / df.groupby('ID')['weight'].transform('sum')
df
```

```python
df['diff_weight'] = abs(df['annotation_utt_score'] - df['weighted_average'])
df
```

```python
weight_diff = df['diff_weight'].mean()

print(f"Average difference with mean: {weight_diff}")
# Worse method so far
```
