# Sentence sentiment score aggregation - mean and median

Testing different aggregation methods (average, mean, etc), as well as investigating some key characteristics of the dataset (e.g., dispersion of the sentence sentiment scores within the utterance). 

```python
aggregation = df.groupby('ID')['annotation_sent'].agg(
    count='count',
    mean ='mean', 
    median='median',
    min='min', 
    max='max', 
    std='std',
).reset_index()
```

```python
df = pd.merge(df, aggregation, on='ID', how='left')
df.head()
```

```python
#Checking the highest data dispersion by sorting the values by standard deviation score 
# count = df.sort_values(by = 'std', ascending=False)
#count.head()
```

```python
df['diff_mean'] = abs(df['annotation_utt_score'] - df['mean'])
df['diff_median'] = abs(df['annotation_utt_score'] - df['median'])

df.head()
```

```python
average_max = df.sort_values(by = 'diff_mean', ascending=False)
average_max.head(5)
```

```python
median_max = df.sort_values(by ='diff_median', ascending=False)
median_max.head(5)
```

```python
mean_diff = df['diff_mean'].mean()
median_diff = df['diff_median'].mean()

print(f"Average difference with mean: {mean_diff}")
print(f"Average difference with median: {median_diff}")
```

## Calculating differents weights and weighted average

The weights assigned determined by the following indicators:
<!-- This is currently just a test with contexts that could be important when considering sentiment in whole utterances -->
- the *position* of the sentence (e.g., if the sentence appears at the end of the utterance)
- the *length* of the sentences (the longer the sentence, the higher the weight?)
- lower weights for the *procedural* sentences (maybe based on the “Chair” value for the utterance to signify high possibility of procedural language in sentences).
    - e.g. *if chair = True & (value ≠ Neutral_P |  value ≠ Neutral_N ) → lower weights set*

```python
df['position'] = df.groupby('ID').cumcount() + 1
#df['position_weight'] = df.groupby('ID')['position'].transform(lambda x: x.apply(lambda y: 1.5 if y == x.max() else 1 ))
df.head(3)
```

```python
df['position_max'] = df.groupby('ID')['position'].transform('max')
df.head()
```

```python
def position_chair(row):
    chair = row['chair']
    position = row['position']
    max_position = row['position_max']
    if chair != True and position == max_position:
        return 1.5
    return 1

df['weight'] = df.apply(position_chair, axis=1)
df.drop(columns=['position_max'], inplace=True)
df.head(10)
```

```python
df['weighted_score'] = df['annotation_sent'] * df['weight']
df['weighted_average'] = df.groupby('ID')['weighted_score'].transform('sum') / df.groupby('ID')['weight'].transform('sum')
df.head()
```

```python
df['diff_weight'] = abs(df['annotation_utt_score'] - df['weighted_average'])
df
```

```python
weight_diff = df['diff_weight'].mean()

print(f"Average difference with chair/position mean: {weight_diff}")
# Worse method so far
```

## Length-based aggregation

```python
df['length'] = df['text_sent'].apply(lambda x: len(x.split()))
```

```python
aggregated_scores = (
    df.groupby("ID")
    .apply(lambda x: (x["annotation_sent"] * x["length"]).sum() / x["length"].sum())
    .reset_index(name="length_aggregated_score")
)
```

```python
df = df.merge(
    aggregated_scores,
    on="ID"
)

df.head()
```

```python
df['diff_length'] = abs(df['annotation_utt_score'] - df['length_aggregated_score'])
length_avg = df['diff_length'].mean()
print("Average difference with length weights :", length_avg)
```
