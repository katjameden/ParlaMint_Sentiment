## Sentiment trends

````python
parties = df['Speaker_party'].value_counts()
orientations = df['Party_orientation'].value_counts()
status = df['Party_status'].value_counts()

percent_parties = (parties / len(df)) * 100
percent_orientations = (orientations / len(df)) * 100
percent_staus = (status / len(df)) * 100

print("% of speeches per political party:", percent_parties)
print("% of speeches for coalition/opposition: ", percent_staus)
print("% of speeches per political orientation:", percent_orientations)
````

````python
status = df.groupby(['Speaker_party','Party_status']).size().unstack(fill_value=0)
status.columns = ['Total Speeches (Unknown)', 'Total Speeches (Coalition)', 'Total Speeches (Opposition)']

avg_sentiment = df.groupby(['Speaker_party', 'Party_status'])['u_annotations'].mean().unstack(fill_value=0)
avg_sentiment.columns = ['Avg. Sentiment (Unknown)', 'Avg. Sentiment (Coalition)', 'Avg. Sentiment (Opposition)']

sentiment_counts = df.groupby(['Speaker_party', df['u_sentiment']]).size().unstack(fill_value=0)
sentiment_percentages = sentiment_counts.div(sentiment_counts.sum(axis=1), axis=0) * 100

party_metadata = df[['Speaker_party', 'Party_orientation']].drop_duplicates().set_index('Speaker_party')

overview = (
    status
    .merge(avg_sentiment, on='Speaker_party', how='left')
    .merge(sentiment_percentages, on='Speaker_party', how='left')
    .reset_index()
)

overview = overview.merge(party_metadata, on='Speaker_party', how='left')

overview.rename(columns={
    'Positive': '% Positive',
    'Neutral': '% Neutral',
    'Negative': '% Negative'
}, inplace=True)

overview['Total Speeches'] = overview['Total Speeches (Unknown)'] + overview['Total Speeches (Coalition)'] + overview['Total Speeches (Opposition)']
total_speeches = overview['Total Speeches'].sum()
overview['Speech %'] = (overview['Total Speeches'] / total_speeches) * 100
overview = overview[['Speaker_party', 'Party_orientation', 'Total Speeches', 'Speech %', 'Total Speeches (Coalition)', 'Total Speeches (Opposition)','Total Speeches (Unknown)',
               'Avg. Sentiment (Coalition)', 'Avg. Sentiment (Opposition)','Avg. Sentiment (Unknown)', '% Positive', '% Neutral', '% Negative']]

overview['Speech %'] = overview['Speech %'].apply(lambda x: f'{x:.2f}')
overview['Avg. Sentiment (Coalition)'] = overview['Avg. Sentiment (Coalition)'].apply(lambda x: f'{x:.2f}')
overview['Avg. Sentiment (Opposition)'] = overview['Avg. Sentiment (Opposition)'].apply(lambda x: f'{x:.2f}')
overview['Avg. Sentiment (Unknown)'] = overview['Avg. Sentiment (Unknown)'].apply(lambda x: f'{x:.2f}')
overview['% Positive'] = overview['% Positive'].apply(lambda x: f'{x:.2f}')
overview['% Neutral'] = overview['% Neutral'].apply(lambda x: f'{x:.2f}')
overview['% Negative'] = overview['% Negative'].apply(lambda x: f'{x:.2f}')



overview
````

#### Party activity, ...

Checking party activity and other trends that allow for a good term characteristics overview.
For these analyses, the question of how to treat "non-parties" (NeP and NP) and anomalys (such as Lipa, Konkretno) are directly tied to visualisation and data granularity. 
- NeP = parliamentary group of independent MPs, which could be considered for the analysis, however, since there is no destinction between different NePs in different mandates, this should be ruled out
- NP = individual independent MPs, not connected into a parliamentary group. In addition to them not being a formal parliamentary entity but individual speakers, they have similar problem as NeP, ruled out
- Lipa and Konkretno = both are non-parliamentary parties. Specfically, both parties were consistuted within Parliament after being elected as a different parliamentary party. Therefore, the party was never directly voted into the Parliament, however, individual MPs, that were elected into the Parliament are still MPs, but within a new party. Since they did produce some speeches within their (very short) terms, they should be kept in the analysis.
- '-' = individual speakers without formal political party affiliation. As we are interested in party activity, they will not be a part of the analysis.
- Some parties have been renamed during the course of their existence within the Parliament, and are in essence a direct continuation of one another (ParlaMint-SI listOrg ```<listEvent>```):
    - SDS.1 (Socialdemokratska stranka Slovenije) -> SDS.2 (Slovenska demokratska stranka)
    - ZL (Levica.1) -> Levica (Levica.2)
    - ZaAB -> ZaSLD -> SAB
    - ZLSD -> SD
    - SLS+SKD -> SLS
    - DLGV -> DL
    - SMC.1 (Stranka Mira Cerarja) -> SMC.2 (Stranka modernega centra) + GAS -> Konkretno (SMC/GAS/Konkretno)

Two versions of the Gantt plot: 
- Visualising party activity with parties as they are.
- Visualising party activity with grouped parties
- Ideally, both plots can be used (with different colour palette) within the Chapter

````python
# Groups of political parties, as per <listEvent> 
groups = {
    "ZL": "ZL/Levica",
    "Levica": "ZL/Levica",
    "ZaAB":"ZaAB/ZaSLD/SAB",
    "ZaSLD":"ZaAB/ZaSLD/SAB",
    "SAB" :"ZaAB/ZaSLD/SAB",
    "ZLSD":"ZLSD/SD",
    "SD":"ZLSD/SD",
    "SLS+SKD":"SLS+SKD/SLS",
    "SLS":"SLS+SKD/SLS", 
    "DLGV":"DLGV/DL", 
    "DL":"DLGV/DL",
    "SMC":"SMC/GAS/Konkretno",
    "Konkretno": "SMC/GAS/Konkretno"

}
````

````python
activity = df.groupby(['Term', 'Speaker_party']).agg(
    Start_date=('Date', 'min'),
    End_date=('Date', 'max'),
    Speech_count=('Date', 'size')
).reset_index()
activity = activity[~activity['Speaker_party'].isin(["-", "NeP", "NP"])]

activity["Grouped_parties"] = activity['Speaker_party'].replace(groups).fillna(activity['Speaker_party'])
activity.head()
````

Analysis of the party activity: By producing both party activity with the unaltered political parties (Figure A) and grouped political parties based on added context (Figure B) and their presence in the Parliament, annotated directly in the (TEI and TEI.ana versions of) corpus as ```<listEvent>```,
 we are able to distinguish several important points: 
- Figure A shows each political party where we consider each party as a separate entity and its standalone contribution in the parliament, 
which gives us a fair overvier. However, using external resources (such as mandate reports), we are able to see that **some** of the parties 
that seem to exist for a limited period are in fact part of a much larger parties under different names. 
- Example of such parties are for examples, parties ZLSD and SD - in Figure A, ZLSD exist for only almost four years (Nov 2002 - Oct 2004). 
However, by combining them with their renamed party, SD, in Figure B (ZLSD/SD), we are able to see that the party has in fact been 
(under similar constitution) present and active in Parliament for the entire duration (or rather, period covered). This makes these party one of the three parties that were present 
in the Parliament for the entire period covered (the other two begin SDS and DeSUS).
- A similar example are parties ZaAB, ZaSLD and SAB, which all represent one party that underwent two renamings during their presence in the Parliament, 
starting as ZaAB (Zavezništvo Alenke Bratušek) in the beginning (2014), with the first renaming to ZaSLD (Zavezništvo socialno-liberalnih demokratov), 
which is not present in Figure A, as they were present in the Parliament, however, due to two MPs from the ZaAB PS, the no. of MPs was too low to be considered a full parliamentary group, and Alenka Bratušek (the president of the parliamentary group) moved on to join the independent MPs which was later constituted as a parliamentary group of independent MPs. During this period, the party was renamed and was not considered a full parliamentary party (Državni zbor, Mandatno poročilo 2014-2018). therefore, no speeches were ever produced by them, which is seen in Figure B as a gap between ZaAB and SAB party activity, with SAB being the last renaming of the party (footnote: However, the party was in Term 9 merged with the current biggest coalition party, Svoboda). 

The grouping of parties shows us a more clear and real picture of party activity - however, there are some cases, where a distinction needs to be made. Such is a case of SMC and Konkretno, which are both in Figure A and B considered as a separate entity. Reason for this is the fact that Konkretno, while active in the Parliament, was never considered an explicit parliamentary party/group, but has been merged with another non-parliamentary party to produce Konkretno, a coalition of parties, not voted into the Parliament. 

In terms of the party activity, as already seen in Table (party characteristics), the undisputed champion in terms of no. of speeches produced is SDS, Term 5 (2008 - 2011) being one of the most active legislative period, with SD and LDS following closely. In more general terms, larger political parties (such as SDS, SD, NSi) tend to be more vocal, which is not surprising given the share of MPs in the Parliament.
