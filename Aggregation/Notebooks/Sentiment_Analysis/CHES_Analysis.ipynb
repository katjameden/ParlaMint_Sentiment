```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('../../Datasets/Sentiment_Analysis/CHES-SI.tsv', delimiter='\t', encoding='utf-8')
df
```

```python
#Convert CHES scores to interger
exclude = ['country', 'pm_id', 'ches_id', 'chesversion']
df[df.columns.difference(exclude)] = df[df.columns.difference(exclude)].apply(pd.to_numeric, errors='coerce')
df.head()
```

## LRGEN Scores by Party across Years

```python
lrgen = df.groupby(['pm_id', 'year'])[['lrgen']].mean().reset_index()
lrgen['year'] = lrgen['year'].astype(int)
lrgen.head()
```

```python
fig = px.line(lrgen, x='year', y='lrgen', color='pm_id', 
              markers=True, title='LRGEN Scores by Parties Across Years',
              labels={'year': 'Year', 'lrgen': 'CIVLIB Score', 'pm_id': 'Party'})

# Improve layout
fig.update_layout(
    xaxis=dict(tickmode='linear', dtick=3),  # Step of 3 for x-axis ticks
    yaxis_title='LRGEN Score',
    legend_title='Party',
    template='plotly_white'
)
```
