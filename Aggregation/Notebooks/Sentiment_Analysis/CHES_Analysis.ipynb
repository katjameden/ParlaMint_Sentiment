```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('../../Datasets/Sentiment_Analysis/CHES-SI.tsv', delimiter='\t', encoding='utf-8')
df
```

```python
#Convert CHES scores to interger
exclude = ['country', 'pm_id', 'ches_id', 'chesversion']
df[df.columns.difference(exclude)] = df[df.columns.difference(exclude)].apply(pd.to_numeric, errors='coerce')
df.head()
```

```python
groups = {
    "ZL": "ZL/Levica",
    "Levica": "ZL/Levica",
    "ZaAB":"ZaAB/ZaSLD/SAB",
    "ZaSLD":"ZaAB/ZaSLD/SAB",
    "SAB" :"ZaAB/ZaSLD/SAB",
    "ZLSD":"ZLSD/SD",
    "SD":"ZLSD/SD",
    "SLS+SKD":"SLS+SKD/SLS",
    "SLS":"SLS+SKD/SLS", 
    "DLGV":"DLGV/DL", 
    "DL":"DLGV/DL",
    "SMC":"SMC/GAS/Konkretno",
    "Konkretno": "SMC/GAS/Konkretno"

}
df["grouped_party"] = df['pm_id'].replace(groups).fillna(df['pm_id'])
df = df.sort_values(by='year')
df
```

## LRGEN Scores by Party across Years

```python
lrgen = df.dropna(subset=["lrgen"])
fig = px.line(lrgen, x='year', y='lrgen', color='grouped_party', 
              markers=True, title='LRGEN Scores by Parties Across Years',
              labels={'year': 'Year', 'lrgen': 'LRGEN Score', 'grouped_party': 'Party'}, color_discrete_sequence=px.colors.qualitative.Vivid,
)


fig.update_layout(
    height=650,  
    width=1000,
    xaxis=dict(tickmode='linear', dtick=3),
    yaxis_title='LRGEN Score',
    legend_title='Party',
    template='plotly_white'
)

fig.write_image('../../Figures/Sentiment_Analysis/CHES/LRGEN.png')
fig.show()
```

# Specific EU Policy Questions

## EU EP
Position of the party leadership in YEAR on the (strengthening) powers of the European Parliament. 

Scale: 1 = Strongly opposed, 2 = Opposed, 3 = Somewhat opposed, 4 = Neutral, 5 = Somewhat in favor, 6 = In favor, 7 = Strongly in favor
Scale interpretation: 
- 1 = Strongly opposed → Party leadership is firmly against expanding EP powers.
- 2 = Opposed → Generally against giving more power to the EP.
- 3 = Somewhat opposed → Leans against EP empowerment but not strongly.
- 4 = Neutral → No clear stance or balanced view.
- 5 = Somewhat in favour → Mildly supports strengthening EP powers.
- 6 = In favour → Generally supports expanding the EP's role.
- 7 = Strongly in favour → Fully supports a stronger European Parliament.

## Most Important Issues
Three most important issues for the party over the course of the YEAR (Asked only in **2014**) - MIP_ONE, MIP_TWO, MIP_THREE

Numerical values in the data corespond to the topics available in the ***1999-2019 Chapel Hill Expert Survey (CHES) trend file Codebook*** (https://www.chesdata.eu/ches-europe)

```python
mips = ['pm_id', 'grouped_party', 'mip_one', 'mip_two', 'mip_three']
mip = df[mips]
mip = mip.dropna(subset=['mip_one', 'mip_two', 'mip_three']).reset_index()
mip = mip.drop(columns='index')
mip[['mip_one', 'mip_two', 'mip_three']] = mip[['mip_one', 'mip_two', 'mip_three']].astype(int)
mip
```

```python
mip_mapping = {
    1: "Anti-elite rhetoric",
    2: 'Civil liberties',
    3: "Corruption",
    4 :'Decentralization',
    5: "Deregulation",
    6: 'Environment',
    7: "Ethnic minorities",
    8: 'EU integration',
    9: "Immigration",
    10: 'International security',
    11: "Multiculturalism",
    12: 'Nationalism',
    13: 'Public services vs taxes', 
    14: 'Redistribution',
    15: 'Religious principles',
    16: 'Social lifestyle',
    17: 'State intervention',
    18: 'Urban vs. rural',
    29: 'Tie: Corruption and Social lifestyle',
    33: 'Tie: Deregulation and Public services vs taxes',
    42: 'Tie: Public services vs taxes and State intervention'
}
```

```python
mip["issues"] = mip.apply(lambda row: "; ".join(
    filter(None, [
        mip_mapping.get(row["mip_one"], ""),  
        mip_mapping.get(row["mip_two"], ""),
        mip_mapping.get(row["mip_three"], "")
    ])
), axis=1) 

mip
```

```python
mip[["mip_one", "mip_two", "mip_three"]] = mip[["mip_one", "mip_two", "mip_three"]].replace(mip_mapping)
mip = mip.drop(columns=['grouped_party', 'issues'])
mip
```
