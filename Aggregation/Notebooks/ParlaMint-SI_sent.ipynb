```python
import pandas as pd
import csv
```

```python
def process_chunk(df_chunk):
    print(df_chunk.head())
def process_jsonl(file_path, chunksize=1000):
    df_list=[]
    for df_chunk in pd.read_json(file_path, lines=True, chunksize=chunksize):
        df_list.append(df_chunk)
    
    full_df = pd.concat(df_list, ignore_index=True)
    return full_df

file_path = '../../ParlaSent-SI/ParlaMint-SI.meta.jsonl'
df = process_jsonl(file_path, chunksize=1000)
df.head()
```

```python
df = df.drop(columns={'text_en', 'metadata'})
df = df.rename(columns={'logit':'annotations', 'newdoc id':'utt_id'})
df.head()
```

```python
#Dataset for next step: calcuation of the aggregations of the whole ParlaMint-SI
#Added to gitignore, as file is too large.
df.to_csv('../Datasets/SI_sent_development.csv', index=False)
```
