```python
import pandas as pd
import os
```

```python
#Input files - ParlaMint-SI (curl from CLARIN.SI repository)
root_directory = "../../ParlaMint-SI/ParlaMint-SI.txt"
```

```python
ids = []
texts = []
```

```python
for root, _, files in os.walk(root_directory):
    for filename in files: 
        if filename.endswith(".txt") and filename != "00README.txt":
            filepath = os.path.join(root, filename)
            with open(filepath, 'r', encoding = 'utf-8') as file:
                for line in file:
                    if '\t' in line:  # Ensure the line contains a tab character
                        parts = line.strip().split('\t')
                        if len(parts) == 2:  # Ensure there are exactly two parts after splitting
                            ids.append(parts[0])
                            texts.append(parts[1])
                    
```

```python
data = pd.DataFrame({
    'ID': ids,
    'Text': texts
})
```
