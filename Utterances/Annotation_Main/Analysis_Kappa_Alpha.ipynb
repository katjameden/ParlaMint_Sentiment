## Krippendorfs Alpha

### 6-level full 

```python
import krippendorff
import numpy as np
```

```python
columns_to_keep = ['tag_tamara', 'tag_katja']
alpha = df[columns_to_keep]
alpha
```

```python
alpha = alpha.dropna()
len(df)
```

```python
unique_labels = pd.unique(alpha[['tag_tamara', 'tag_katja']].values.ravel())

# Create a mapping of labels to consistent numeric codes
label_to_code = {label: idx for idx, label in enumerate(unique_labels)}

# Map the labels to numeric codes for both annotators
alpha['tag_tamara_codes'] = df['tag_tamara'].map(label_to_code)
alpha['tag_katja_codes'] = df['tag_katja'].map(label_to_code)
alpha
```

```python
data_array = alpha[['tag_tamara_codes', 'tag_katja_codes']].to_numpy().T
alpha_calc = krippendorff.alpha(reliability_data=data_array, level_of_measurement='nominal')
print(f"Krippendorff's alpha (nominal): {alpha_calc}")
```

## Aggregation rates

### 6-level

```python
#Add new column 'agreement' with value True if there was agreement between the annotators (i.e., if both annotators used the same label).
df['agreement'] = df['tag_tamara'] == df['tag_katja']
agreement = df[df['agreement'] == True]
len_agreement = len(agreement)
len_agreement
```

```python
agreement_percentage = (len_agreement / len(df)) * 100
agreement_percentage
```

### 6-level non procedural

```python
non = df[df['chair'] == False]
non.head()
len(non)
```

```python
non_agreement = non[non['agreement'] == True]
non_agreement.head()
len(non_agreement)
```

```python
non_agree_percent = (len(non_agreement) / len(non)) * 100
non_agree_percent
```
