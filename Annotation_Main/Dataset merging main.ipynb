# Dataset merge and prep for analysis

Checking and renaming the datasets to merge them into one combined dataset and get it ready for (IAA) analysis.

Change folder (Test1, Test2, Main1, Main2...) to process other datasets as they come.

```python
df1 = pd.read_excel("Main2/Main2_Tamara.xlsx")
df2 = pd.read_excel("Main2/Main2_Katja.xlsx")
```

```python
df1 = df1.drop(columns=['Annotation values', 'Chair'])
df2 = df2.drop(columns='Annotation values')
```

```python
df1.rename(columns={'Text':'text', 'Annotation': 'tag_tamara', 'Comments': 'comments_tamara','Flag':'flagged_tamara', 'Chair':'chair'}, inplace=True)
df1.head()
```

```python
df2.rename(columns={'Text':'text', 'Annotation': 'tag_katja', 'Comments': 'comments_katja','Flag':'flagged_katja', 'Chair':'chair'}, inplace=True)
df2.head()
```

```python
merged_df = pd.merge(df1, df2, on=['ID', 'text'])
merged_df.head()
```

```python
order = ['ID', 'text', 'tag_tamara', 'tag_katja', 'comments_tamara', 'comments_katja', 'flagged_tamara', 'flagged_katja', 'chair']
```

```python
merged_df = merged_df[order]
merged_df['final_tag'] = ''
merged_df.head()
```

```python
merged_df.to_csv("Main2/Main2_combined.csv", encoding ='utf-8', index=False)
```
