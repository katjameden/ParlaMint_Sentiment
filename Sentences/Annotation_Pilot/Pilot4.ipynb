```python
import pandas as pd
import csv

df = pd.read_csv("pilot4.csv", delimiter = ";")
df
```

```python
df1 = df.drop(['timestamp_tamara', 'timestamp_anze', 'timestamp_katja', 'rev_tamara', 'rev_anze', 'rev_katja', 'final_tag'], axis = 1)
df1.head() 
```

```python
from sklearn.metrics import cohen_kappa_score

annotator1 = df1['tag_tamara']
annotator2 = df1['tag_katja']
annotator3 = df1['tag_anze']


cohen_kappa_score(annotator1, annotator2)#Tamara-Katja
```

```python
cohen_kappa_score(annotator2, annotator3)#Katja - Anže
```

```python
cohen_kappa_score(annotator1, annotator3)#Tamara-Anže
```

```python
annotator1 = df1['tag_tamara']
annotator2 = df1['tag_katja']
annotator3 = df1['tag_anze']


kappa1 = cohen_kappa_score(annotator1, annotator2)#Tamara-Katja
kappa2 = cohen_kappa_score(annotator2, annotator3)#Katja-Anže
kappa3 = cohen_kappa_score(annotator1, annotator3)#Tamara-Anže
```

```python
Cohen_kappa = (kappa1 + kappa2 + kappa3)/3
Cohen_kappa
```
