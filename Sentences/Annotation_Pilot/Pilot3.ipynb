```python
import pandas as pd
import csv

sentiment = pd.read_csv("pilot3.csv", delimiter=";")
sentiment.head()
```

```python
df = sentiment.drop(['timestamp_tamara', 'timestamp_anze', 'timestamp_katja', 'rev_tamara', 'rev_anze', 'rev_katja', 'final_tag'], axis = 1)
df.head()
```

```python
df.loc[df['tag_tamara'] == 'N_Neutral', 'sent_tamara'] = 'Neutral'
df.loc[df['tag_tamara'] == 'P_Neutral', 'sent_tamara'] = 'Neutral'
df.loc[df['tag_tamara'] == 'M_Negative', 'sent_tamara'] = 'Negative'
df.loc[df['tag_tamara'] == 'M_Positive', 'sent_tamara'] = 'Positive'
df.loc[df['tag_tamara'] == 'Negative', 'sent_tamara'] = 'Negative'
df.loc[df['tag_tamara'] == 'Positive', 'sent_tamara'] = 'Positive'
```

```python
df.loc[df['tag_katja'] == 'N_Neutral', 'sent_katja'] = 'Neutral'
df.loc[df['tag_katja'] == 'P_Neutral', 'sent_katja'] = 'Neutral'
df.loc[df['tag_katja'] == 'M_Negative', 'sent_katja'] = 'Negative'
df.loc[df['tag_katja'] == 'M_Positive', 'sent_katja'] = 'Positive'
df.loc[df['tag_katja'] == 'Negative', 'sent_katja'] = 'Negative'
df.loc[df['tag_katja'] == 'Positive', 'sent_katja'] = 'Positive'
```

```python
df.loc[df['tag_anze'] == 'N_Neutral', 'sent_anze'] = 'Neutral'
df.loc[df['tag_anze'] == 'P_Neutral', 'sent_anze'] = 'Neutral'
df.loc[df['tag_anze'] == 'M_Negative', 'sent_anze'] = 'Negative'
df.loc[df['tag_anze'] == 'M_Positive', 'sent_anze'] = 'Positive'
df.loc[df['tag_anze'] == 'Negative', 'sent_anze'] = 'Negative'
df.loc[df['tag_anze'] == 'Positive', 'sent_anze'] = 'Positive'
```

```python
from sklearn.metrics import cohen_kappa_score
annotator1 = df['sent_tamara']
annotator2 = df['sent_katja']
annotator3 = df['sent_anze']
cohen_kappa_score(annotator2, annotator3)
```

```python
from sklearn.metrics import cohen_kappa_score
annotator1 = df['tag_tamara']
annotator2 = df['tag_katja']
annotator3 = df['tag_anze']
cohen_kappa_score(annotator2, annotator3)
```
