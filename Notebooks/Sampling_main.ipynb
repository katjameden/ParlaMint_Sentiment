## Sampling main annotation subsets
To avoid duplicates between the tests and the main annotation dataset, the main annotation dataset was resampled by sampling every 309th utterance (in the notebook Sampling_tests.ipynb) to better match the corpus distribution. The order of the utterances was then randomised. This is later used to generate the dev and train split used to test the sentiment score aggregation methods.

For the main annotation (which was split into 5 parts), each main annotation sample comprises 200 utterances to be annotated.

```python
df = pd.read_csv("../Datasets/Sample.csv")
df.head()
```

## Checking the distribution of speech types (procedural/non-procedural)
The dataset is relatively balanced with respect to the classification of speech types.

# Dev and test split for sentiment aggregation

Generating the dev and test split of the new main annotation sample that will be used to develop and test method for the sentiment score aggregation.


```python
from sklearn.model_selection import train_test_split
dev_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
print(f'Development set size: {len(dev_df)} rows')
print(f'Test set size: {len(test_df)} rows')
```

```python
dev_df.to_csv('../Datasets/Sample_dev.csv', index=False)
test_df.to_csv('../Datasets/Sample_test.csv', index=False)
```
