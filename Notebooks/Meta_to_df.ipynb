```python
import os
import glob
import pandas as pd
```

```python
#Input files - ParlaMint-SI (curl from CLARIN.SI repository)
input_directory = "../../ParlaMint-SI/ParlaMint-SI.txt"
output_directory = "../"
output_file = "../ParlaMint-SI-meta.csv"
```

```python
if os.path.exists(output_directory):
    os.remove(output_file)
```

```python
meta_files = glob.glob(os.path.join(input_directory, '**', '*-meta.tsv'), recursive=True)
```

```python
data = []
header_written = False
with open(output_file, 'w') as outfile:
    for file in meta_files:
        with open(file, 'r') as infile:
            header = infile.readline()
            if not header_written:
                outfile.write(header)
                data.append(header.strip().split('\t'))  # Add header to data list
                header_written = True
            for line in infile:
                outfile.write(line)
                data.append(line.strip().split('\t'))  # Add data to list
                
print(f"Combined metadata written to {output_file}")
```

```python
df = pd.DataFrame(data[1:], columns=data[0])

df

# Optionally, save the DataFrame to a CSV file (if needed)
#df.to_csv('../Parlamint-SI-meta.csv', index=False)
```
